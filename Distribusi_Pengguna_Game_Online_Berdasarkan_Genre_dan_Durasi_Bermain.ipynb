```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data pengguna game online berdasarkan genre dan durasi bermain (jam per minggu)
data = {
    'Genre': [
        'MOBA', 'FPS', 'MMORPG', 'Battle Royale', 'Simulation',
        'Racing', 'Strategy', 'Sports', 'Puzzle', 'Horror'
    ],
    'Pengguna (juta)': [35, 40, 25, 30, 20, 15, 18, 22, 10, 8],
    'Durasi Rata-rata (jam/minggu)': [12, 15, 20, 14, 10, 9, 13, 11, 6, 7]
}

df = pd.DataFrame(data)
plt.figure(figsize=(12, 6))
sns.set(style='whitegrid')
ax = sns.barplot(
    data=df,
    x='Genre',
    y='Pengguna (juta)',
    palette='viridis'
)

for i, row in df.iterrows():
    ax.text(
        i,
        row['Pengguna (juta)'] + 1,
        f"{row['Durasi Rata-rata (jam/minggu)']} jam/mgg",
        ha='center',
        va='bottom',
        fontsize=9
    )

plt.title('Jumlah Pengguna Game Online per Genre (dengan Durasi Rata-rata Bermain)')
plt.xlabel('Genre')
plt.ylabel('Jumlah Pengguna (juta)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
