```python
# baca file exel

import pandas as pd

df = pd.read_excel("data_kuesioner_pelanggaran_lalu_lintas_100.xlsx")
df_clean = df.drop(columns=["ID"])  # Buang kolom ID
df_clean.head()
```

```python
# visualisasi histogram
import matplotlib.pyplot as plt
import seaborn as sns

# Buat histogram untuk setiap kolom
plt.figure(figsize=(15, 10))
for i, col in enumerate(df_clean.columns, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df_clean[col], bins=5, kde=True, color='lightcoral')
    plt.title(f'Histogram: {col}')
    plt.xlabel('Skor')
    plt.ylabel('Frekuensi')
    plt.xticks(range(1, 6))

plt.tight_layout()
plt.show()
```
